Importing relevant modules and libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Loading data and have a look through the data

```python
df = pd.read_csv("all_data.csv")
print(df.Country.unique())
print(df.Year.unique())
df.rename(columns = {"Life expectancy at birth (years)": "Life expectancy"}, inplace = True)
print(df.head())
```

Question 1: 
Has life expectancy increased over time in the six nations?

```python
chile = df[df.Country == "Chile"]
china = df[df.Country == "China"]
germany = df[df.Country == "Germany"]
mexico = df[df.Country == "Mexico"]
usa = df[df.Country == "United States of America"]
zimbabwe = df[df.Country == "Zimbabwe"]

plt.figure(figsize = (12,7))
ax1 = plt.subplot(2, 3, 1)
ax1.plot(chile.Year, chile["Life expectancy"], label = "Chile")
ax1.set_ylabel("Life expectancy at birth (years)")
plt.legend()

ax2 = plt.subplot(2, 3, 2)
ax2.plot(china.Year, china["Life expectancy"], label = "China")
plt.legend()

ax3 = plt.subplot(2, 3, 3)
ax3.plot(germany.Year, germany["Life expectancy"], label = "Germany")
plt.legend()

ax4 = plt.subplot(2, 3, 4)
ax4.plot(mexico.Year, mexico["Life expectancy"], label = "Mexico")
ax4.set_ylabel("Life expectancy at birth (years)")
ax4.set_xlabel("Year")
plt.legend()

ax5 = plt.subplot(2, 3, 5)
ax5.plot(usa.Year, usa["Life expectancy"], label = "United States of America")
ax5.set_xlabel("Year")
plt.legend()

ax6 = plt.subplot(2, 3, 6)
ax6.plot(zimbabwe.Year, zimbabwe["Life expectancy"], label = "Zimbabwe")
ax6.set_xlabel("Year")
plt.legend()

plt.show()
```

```python
# So we can see that the life expectancy of these 6 nations have generally increased over time 
```

Question 2: Has GDP increased over time in the six nations?


```python
plt.figure(figsize = (12,7))
ax1 = plt.subplot(2, 3, 1)
ax1.plot(chile.Year, chile["GDP"], label = "Chile")
ax1.set_ylabel("GDP")
plt.legend()

ax2 = plt.subplot(2, 3, 2)
ax2.plot(china.Year, china["GDP"], label = "China")
plt.legend()

ax3 = plt.subplot(2, 3, 3)
ax3.plot(germany.Year, germany["GDP"], label = "Germany")
plt.legend()

ax4 = plt.subplot(2, 3, 4)
ax4.plot(mexico.Year, mexico["GDP"], label = "Mexico")
ax4.set_ylabel("GDP")
ax4.set_xlabel("Year")
plt.legend()

ax5 = plt.subplot(2, 3, 5)
ax5.plot(usa.Year, usa["GDP"], label = "United States of America")
ax5.set_xlabel("Year")
plt.legend()

ax6 = plt.subplot(2, 3, 6)
ax6.plot(zimbabwe.Year, zimbabwe["GDP"], label = "Zimbabwe")
ax6.set_xlabel("Year")
plt.legend()

plt.show()
```

```python
# The GDP of these six nations has also increased though we can see that for some countries like Germany, Chile and Mexico, 
# the GDP appears to be plateauing or even beginning to fall.
```

Question 3: Is there a correlation between GDP and life expectancy of a country?

```python
plt.figure(figsize = (12,7))
sns.lmplot(data = chile, x = "Life expectancy",y = "GDP")
ax1.set_xlabel("Life expectancy at birth (years)")
plt.legend(["Chile"])
plt.show()
plt.clf()

sns.lmplot(data = china, x = "Life expectancy",y = "GDP")
plt.legend(["China"])
plt.show()
plt.clf()

sns.lmplot(data = germany, x = "Life expectancy",y = "GDP")
plt.legend(["Germany"])
plt.show()
plt.clf()

sns.lmplot(data = mexico, x = "Life expectancy",y = "GDP")
ax4.set_xlabel("Life expectancy at birth (years)")
ax4.set_ylabel("GDP")
plt.legend(["Mexico"])
plt.show()
plt.clf()

sns.lmplot(data = usa, x = "Life expectancy",y = "GDP")
ax5.set_xlabel("GDP")
plt.legend(["United States of America"])
plt.show()
plt.clf()

sns.lmplot(data = zimbabwe, x = "Life expectancy",y = "GDP")
ax6.set_xlabel("GDP")
plt.legend(["Zimbabwe"])
plt.show()
plt.clf()
```

```python
# There is positive correlation between life expectancy and GDP, ie as life expectancy increases, the GDP will 
# increase as well or vice versa
```

Question 4: What is the average life expectancy in these nations?


```python
countries = ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
ax = plt.subplot()
sns.barplot(data = df, x = "Country", y = "Life expectancy", palette = "Set2")
plt.legend(countries)
plt.xticks(rotation = 40)
ax.set_xticks(range(len(countries)))
ax.set_xticklabels(countries)
plt.show()
```

```python
# The average life expectacny of these countries is generally at the 70 to 80 years range with the exception only of Zimbabwe, 
# which is at about 50. This is probably due to the fact that Zimbabwe is a lot less developed than the other countries.
```
